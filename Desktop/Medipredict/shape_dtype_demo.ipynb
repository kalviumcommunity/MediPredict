## Data Types Interpretation

Each column has a specific data type:

- object → text/string data
- int64 → integer numbers
- float64 → decimal numbers

Correct data types are important for performing operations correctly.

## Dataset Structure Insight

- The dataset contains structured tabular data
- Some numeric-looking columns may be stored as strings
- This can cause errors in calculations

## Scenario Explanation

If a numeric column is stored as a string, mathematical operations will fail.

By checking data types early using df.dtypes, we can detect such issues.

Signals to watch:
- Numbers stored as 'object'
- Unexpected string values in numeric columns

Early inspection helps prevent errors during analysis.
