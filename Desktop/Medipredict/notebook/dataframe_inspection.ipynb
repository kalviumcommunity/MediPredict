# Milestone 4.31 - Pandas DataFrame Inspection

This notebook demonstrates comprehensive inspection of Pandas DataFrames:

- Load a dataset into a Pandas DataFrame
- Inspect DataFrame shape
- Inspect column data types
- Interpret what rows, columns, and types represent
- Clean and readable code

```python
import pandas as pd
import numpy as np

# Create a comprehensive healthcare dataset
np.random.seed(42)  # For reproducible results

# Generate realistic patient data
num_patients = 100
patient_ids = range(1001, 1001 + num_patients)

# Patient demographics
ages = np.random.randint(18, 85, num_patients)
genders = np.random.choice(['Male', 'Female'], num_patients)
names = [f'Patient_{i}' for i in range(1, num_patients + 1)]

# Medical measurements
systolic_bp = np.random.randint(90, 180, num_patients)
diastolic_bp = np.random.randint(60, 120, num_patients)
heart_rates = np.random.randint(50, 120, num_patients)
temperatures = np.round(np.random.uniform(96.0, 101.0, num_patients), 1)

# Hospital information
departments = np.random.choice(['Cardiology', 'Neurology', 'Orthopedics', 'Emergency', 
                                'Pediatrics', 'Oncology'], num_patients)
admission_dates = pd.date_range('2024-01-01', periods=num_patients, freq='H')
length_of_stay = np.random.randint(1, 15, num_patients)

# Treatment and outcomes
treatments = np.random.choice(['Surgery', 'Medication', 'Therapy', 'Observation'], num_patients)
insurance_types = np.random.choice(['Private', 'Medicare', 'Medicaid', 'Self-Pay'], num_patients)
costs = np.round(np.random.uniform(1000, 50000, num_patients), 2)

# Create the DataFrame
healthcare_data = {
    'patient_id': patient_ids,
    'patient_name': names,
    'age': ages,
    'gender': genders,
    'systolic_bp': systolic_bp,
    'diastolic_bp': diastolic_bp,
    'heart_rate': heart_rates,
    'temperature': temperatures,
    'department': departments,
    'admission_date': admission_dates,
    'length_of_stay': length_of_stay,
    'treatment': treatments,
    'insurance_type': insurance_types,
    'total_cost': costs
}

# Load data into DataFrame
df = pd.DataFrame(healthcare_data)

print("Healthcare DataFrame loaded successfully!")
print(f"Dataset created with {len(df)} patient records")
```

## 1. DataFrame Shape Inspection

Examining the dimensions of our dataset to understand its size.

```python
# Get DataFrame shape
shape = df.shape
print(f"DataFrame Shape: {shape}")
print(f"Number of rows: {shape[0]}")
print(f"Number of columns: {shape[1]}")

# Interpretation
print("\n=== INTERPRETATION ===")
print(f"• Rows ({shape[0]}): Each row represents a unique patient record")
print(f"• Columns ({shape[1]}): Each column represents a specific patient attribute or measurement")
print(f"• Total data points: {shape[0] * shape[1]} individual pieces of information")
```

## 2. Column Data Types Inspection

Analyzing the data types of each column to understand the nature of our data.

```python
# Get column data types
print("Column Data Types:")
print("=" * 40)
df_types = df.dtypes
print(df_types)

# Count data types
print("\nData Type Summary:")
print("=" * 40)
type_counts = df_types.value_counts()
for dtype, count in type_counts.items():
    print(f"{dtype}: {count} columns")

# Detailed interpretation
print("\n=== DETAILED INTERPRETATION ===")
print("=" * 40)
for col, dtype in df_types.items():
    if dtype == 'int64':
        print(f"• {col} ({dtype}): Whole numbers - ideal for counting, indexing, mathematical operations")
    elif dtype == 'float64':
        print(f"• {col} ({dtype}): Decimal numbers - perfect for measurements, costs, precise calculations")
    elif dtype == 'object':
        print(f"• {col} ({dtype}): Text/categorical data - names, categories, descriptive information")
    elif dtype == 'datetime64[ns]':
        print(f"• {col} ({dtype}): Date/time data - timestamps, temporal analysis, time series")
```

## 3. Structure Analysis and Interpretation

Understanding what our DataFrame structure represents in a healthcare context.

```python
# Display first few rows to understand structure
print("Sample Data (First 5 rows):")
print("=" * 50)
print(df.head())

# Column categories by type
print("\n=== COLUMN CATEGORIES ===")
print("=" * 30)

# Identify different types of columns
id_columns = ['patient_id']
demographic_columns = ['patient_name', 'age', 'gender']
vital_signs_columns = ['systolic_bp', 'diastolic_bp', 'heart_rate', 'temperature']
administrative_columns = ['department', 'admission_date', 'length_of_stay', 'insurance_type']
clinical_columns = ['treatment']
financial_columns = ['total_cost']

print(f"\n📋 IDENTIFICATION COLUMNS ({len(id_columns)}):")
for col in id_columns:
    print(f"  • {col}: Unique patient identifier")

print(f"\n👥 DEMOGRAPHIC COLUMNS ({len(demographic_columns)}):")
for col in demographic_columns:
    print(f"  • {col}: Patient demographic information")

print(f"\n🏥 VITAL SIGNS COLUMNS ({len(vital_signs_columns)}):")
for col in vital_signs_columns:
    print(f"  • {col}: Medical measurements and health indicators")

print(f"\n📋 ADMINISTRATIVE COLUMNS ({len(administrative_columns)}):")
for col in administrative_columns:
    print(f"  • {col}: Hospital administrative and logistical data")

print(f"\n🩺 CLINICAL COLUMNS ({len(clinical_columns)}):")
for col in clinical_columns:
    print(f"  • {col}: Treatment and medical intervention data")

print(f"\n💰 FINANCIAL COLUMNS ({len(financial_columns)}):")
for col in financial_columns:
    print(f"  • {col}: Cost and billing information")
```

## 4. Detailed Column Inspection

Examining specific columns to understand their data characteristics.

```python
# Detailed inspection of key columns
print("DETAILED COLUMN INSPECTION")
print("=" * 40)

# Numeric columns inspection
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(f"\n🔢 NUMERIC COLUMNS ({len(numeric_cols)}):")
for col in numeric_cols:
    print(f"\n  {col}:")
    print(f"    - Type: {df[col].dtype}")
    print(f"    - Range: {df[col].min()} to {df[col].max()}")
    print(f"    - Mean: {df[col].mean():.2f}")
    print(f"    - Non-null values: {df[col].count()}/{len(df)}")
    
    # Context-specific interpretation
    if 'age' in col:
        print(f"    - Context: Patient age in years")
    elif 'bp' in col:
        print(f"    - Context: Blood pressure measurement in mmHg")
    elif 'heart_rate' in col:
        print(f"    - Context: Heart beats per minute")
    elif 'temperature' in col:
        print(f"    - Context: Body temperature in Fahrenheit")
    elif 'cost' in col:
        print(f"    - Context: Total treatment cost in USD")

# Categorical columns inspection
categorical_cols = df.select_dtypes(include=['object']).columns
print(f"\n📝 CATEGORICAL COLUMNS ({len(categorical_cols)}):")
for col in categorical_cols:
    print(f"\n  {col}:")
    print(f"    - Type: {df[col].dtype}")
    print(f"    - Unique values: {df[col].nunique()}")
    print(f"    - Non-null values: {df[col].count()}/{len(df)}")
    if df[col].nunique() <= 10:  # Show values if not too many
        print(f"    - Values: {list(df[col].unique())}")
    
    # Context-specific interpretation
    if 'gender' in col:
        print(f"    - Context: Patient gender classification")
    elif 'department' in col:
        print(f"    - Context: Medical specialty department")
    elif 'treatment' in col:
        print(f"    - Context: Type of medical treatment provided")

# Date columns inspection
date_cols = df.select_dtypes(include=['datetime64[ns]']).columns
print(f"\n📅 DATE/TIME COLUMNS ({len(date_cols)}):")
for col in date_cols:
    print(f"\n  {col}:")
    print(f"    - Type: {df[col].dtype}")
    print(f"    - Date range: {df[col].min()} to {df[col].max()}")
    print(f"    - Non-null values: {df[col].count()}/{len(df)}")
    print(f"    - Context: Hospital admission timestamp")
```

## 5. Summary and Interpretation

Comprehensive summary of our DataFrame inspection findings.

```python
# Complete DataFrame summary
print("COMPREHENSIVE DATAFRAME SUMMARY")
print("=" * 50)

print(f"📊 DATASET OVERVIEW:")
print(f"   • Total Records: {len(df)} patients")
print(f"   • Total Attributes: {len(df.columns)} features")
print(f"   • Total Data Points: {len(df) * len(df.columns)} values")

print(f"\n📋 DATA DISTRIBUTION:")
for dtype, count in df.dtypes.value_counts().items():
    percentage = (count / len(df.columns)) * 100
    print(f"   • {dtype}: {count} columns ({percentage:.1f}%)")

print(f"\n🔍 DATA QUALITY:")
null_counts = df.isnull().sum()
total_nulls = null_counts.sum()
print(f"   • Missing values: {total_nulls} ({(total_nulls/(len(df)*len(df.columns)))*100:.2f}%)")
print(f"   • Complete records: {len(df) - null_counts.max()}")

print(f"\n🏥 HEALTHCARE CONTEXT:")
print(f"   • This dataset represents patient records from a hospital")
print(f"   • Each row is a unique patient admission/encounter")
print(f"   • Columns capture demographics, vital signs, treatments, and costs")
print(f"   • Suitable for patient outcome analysis, resource planning, and cost studies")

print(f"\n✅ INSPECTION COMPLETE:")
print(f"   • DataFrame structure is well-organized and complete")
print(f"   • Data types are appropriate for their respective columns")
print(f"   • No missing values detected in this synthetic dataset")
print(f"   • Ready for data analysis and visualization tasks")
```
