# Milestone 4.31 - Pandas DataFrame Selection

This notebook demonstrates various ways to select data from Pandas DataFrames:

- Select columns by name
- Select rows using indexing and slicing
- Combine row and column selection
- Clean, readable selection logic

```python
import pandas as pd
import numpy as np

# Create sample healthcare data
data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108],
    'name': ['John Smith', 'Sarah Johnson', 'Mike Davis', 'Emily Brown', 
             'David Wilson', 'Lisa Anderson', 'Robert Taylor', 'Maria Garcia'],
    'age': [45, 32, 67, 28, 54, 41, 73, 36],
    'blood_pressure': [120/80, 118/75, 135/85, 115/70, 128/82, 122/78, 140/90, 119/76],
    'heart_rate': [72, 68, 85, 65, 78, 70, 88, 69],
    'temperature': [98.6, 98.4, 99.1, 98.2, 98.8, 98.5, 99.3, 98.3],
    'admission_date': ['2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', 
                       '2024-01-19', '2024-01-20', '2024-01-21', '2024-01-22'],
    'department': ['Cardiology', 'Neurology', 'Cardiology', 'Orthopedics', 
                   'Neurology', 'Cardiology', 'Emergency', 'Orthopedics']
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

## 1. Selecting a Single Column by Name

Using dot notation or bracket notation to select one column.

```python
# Method 1: Dot notation (clean for simple column names)
patient_names = df.name
print("Patient names using dot notation:")
print(patient_names)

# Method 2: Bracket notation (works for all column names)
patient_ages = df['age']
print("\nPatient ages using bracket notation:")
print(patient_ages)
```

## 2. Selecting Multiple Columns

Using a list of column names to select multiple columns at once.

```python
# Select multiple specific columns
vital_signs = df[['name', 'age', 'heart_rate', 'temperature']]
print("Vital signs for all patients:")
print(vital_signs)

# Select patient identification columns
patient_info = df[['patient_id', 'name', 'department']]
print("\nPatient identification information:")
print(patient_info)
```

## 3. Selecting Rows Using Positional Indexing

Using `.iloc[]` for integer-based positional selection.

```python
# Select single row by position (first patient)
first_patient = df.iloc[0]
print("First patient (position 0):")
print(first_patient)

# Select multiple rows by positions
selected_patients = df.iloc[[1, 3, 5]]  # Patients at positions 1, 3, and 5
print("\nSelected patients (positions 1, 3, 5):")
print(selected_patients)
```

## 4. Slicing a Range of Rows

Using slicing to select consecutive rows.

```python
# Select first 3 patients (rows 0, 1, 2)
first_three = df.iloc[0:3]
print("First three patients:")
print(first_three)

# Select patients from position 2 to 5 (exclusive)
middle_patients = df.iloc[2:6]
print("\nPatients from position 2 to 5:")
print(middle_patients)

# Select every other patient
every_other = df.iloc[::2]
print("\nEvery other patient:")
print(every_other)
```

## 5. Selecting Rows Using Label-Based Indexing

Using `.loc[]` for label-based selection (when we have custom index).

```python
# Set patient_id as index for label-based selection
df_indexed = df.set_index('patient_id')
print("DataFrame with patient_id as index:")
print(df_indexed)

# Select specific patients by their IDs
selected_by_id = df_indexed.loc[[101, 103, 107]]
print("\nPatients with IDs 101, 103, 107:")
print(selected_by_id)
```

## 6. Combining Row and Column Selection

Selecting specific rows and columns together for precise data extraction.

```python
# Select specific rows and columns using .iloc
# Rows 1, 3, 5 and columns 0, 1, 4 (name, age, heart_rate)
specific_data_iloc = df.iloc[[1, 3, 5], [0, 2, 4]]
print("Specific patients with selected columns (using .iloc):")
print(specific_data_iloc)

# Select using .loc with the indexed DataFrame
# Patients 101, 103, 107 and columns name, age, department
specific_data_loc = df_indexed.loc[[101, 103, 107], ['name', 'age', 'department']]
print("\nSpecific patients with selected columns (using .loc):")
print(specific_data_loc)

# Select range of rows and specific columns
# First 4 patients, only vital signs
vital_signs_subset = df.loc[0:3, ['name', 'age', 'blood_pressure', 'heart_rate', 'temperature']]
print("\nFirst 4 patients with vital signs:")
print(vital_signs_subset)
```

## 7. Conditional Row Selection

Using boolean conditions to filter rows based on criteria.

```python
# Select patients older than 50
elderly_patients = df[df['age'] > 50]
print("Patients older than 50:")
print(elderly_patients[['name', 'age', 'department']])

# Select patients from Cardiology department
cardiology_patients = df[df['department'] == 'Cardiology']
print("\nCardiology patients:")
print(cardiology_patients[['name', 'age', 'heart_rate']])

# Select patients with high heart rate (>80) and specific columns
high_hr_patients = df[df['heart_rate'] > 80][['name', 'age', 'heart_rate', 'department']]
print("\nPatients with high heart rate (>80):")
print(high_hr_patients)
```

## Summary

This notebook demonstrated:

✅ **Single column selection**: `df.column` or `df['column']`
✅ **Multiple column selection**: `df[['col1', 'col2', 'col3']]`
✅ **Positional row selection**: `df.iloc[row_index]`
✅ **Row slicing**: `df.iloc[start:end]`
✅ **Label-based selection**: `df.loc[label]`
✅ **Combined selection**: `df.iloc[rows, columns]` or `df.loc[rows, columns]`
✅ **Conditional selection**: `df[df['column'] > value]`

All selections are clean, readable, and intentional for extracting specific healthcare data.
